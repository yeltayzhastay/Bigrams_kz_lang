```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('all_unique_words_2.csv').drop_duplicates()
df
```

```python
df = df.dropna()
df
```

```python
def cleaning_data(code_data):
    text = code_data.kaz.tolist()
    sentences = []
    for line in range(len(text)):
        pre = re.sub(r'http\S+', '', str(text[line])).replace(r'<a href="', '')
        stripped = re.sub('[^a-zA-Z, ^А-Я,а-я,Ә,І,Ң,Ғ,Ү,Ұ,Қ,Ө,Һ,ә,і,ə,ң,ғ,ү,ұ,қ,ө,һ]', ' ', pre.replace('-', ''))
        stripped = stripped.replace('amp', ' ')
        stripped = re.sub('_', '', stripped)
        stripped = re.sub('\s+', ' ', stripped)
        sentences.append(str(stripped).lower())
    return pd.DataFrame(sentences, columns = ['kaz'])
```

```python
df = cleaning_data(df)
df
```

```python
df = df[df.kaz.str.len() > 3]
df
```

```python
df.to_csv('unica.csv', index=False)
```
