```python
import pandas as pd
import numpy as np
import re, codecs
```

```python
def only_arabic(code_data):
    text = code_data.text.tolist()
    label = code_data.label.tolist()
    sentences = []
    for line in range(len(text)):
        stripped = str(" ".join(re.findall(r'[\u0600-\u06FF]+', text[line])))
        stripped = re.sub('_', '', stripped)
        stripped = re.sub('\s+', ' ', stripped)
        if(stripped != ''):
            sentences.append([str(stripped).lower(), int(label[line])])
    return pd.DataFrame(sentences, columns = ['text', 'label'])
```

```python
df = pd.read_csv('Corpus/big_corpus.csv')
df
```

```python
df['count'] = df['text'].str.count(' ') + 1
count = df.loc[(df['label'] == 0) | (df['label'] == 1)]['count'].sum()
print(count)
print(df['count'].sum())
```

```python
only = only_arabic(df)
only
```
