```python
import pandas as pd
from docx2python import docx2python
```

# Extremistical

```python
doc_result_1 = docx2python("Corpus/extr_1.docx")
doc_result_2 = docx2python("Corpus/extr_2.docx")
doc_result_3 = docx2python("Corpus/extr_3.docx")
doc_result_4 = docx2python("Corpus/extr_4.docx")
```

```python
result = []
for i in range(len(doc_result_1.body[0])):
    result.append([doc_result_1.body[0][i][1][0], 1])
for i in range(len(doc_result_2.body[0])):
    result.append([doc_result_2.body[0][i][1][0], 1])
for i in range(len(doc_result_3.body[0])):
    result.append([doc_result_3.body[0][i][1][0], 1])
for i in range(len(doc_result_4.body[0][0][0])):
    if(doc_result_4.body[0][0][0][i] != '' and doc_result_4.body[0][0][0][i] != ' '):
        result.append([doc_result_4.body[0][0][0][i], 1])
```

```python
data_cleaned = []
with open('Corpus/extra.txt','r', encoding="utf-8") as f:
    data_cleaned = [line.replace('\n', '') for line in f]
```

```python
datas = []
iter = -1
for i in range(len(data_cleaned)):
    if(data_cleaned[i] != '' and data_cleaned[i][0].isnumeric()):
        datas.append(data_cleaned[i])
        iter +=1
    else:
        datas[iter] = datas[iter] + data_cleaned[i]
```

```python
for i in range(len(datas)):
    datas[i] = [datas[i], 1]
```

```python
df = pd.DataFrame(result, columns = ['text', 'label'])
```

```python
df
```

```python
old_df = pd.DataFrame(datas, columns = ['text', 'label'])
```

```python
old_df
```

```python
merged = pd.concat([df, old_df], ignore_index=True)
```

```python
merged
```

```python
merged.to_csv('Corpus/extra.csv')
```

```python
from pathlib import Path
def read_docx_folder(path, label):
    all_txt_files =[]
    for file in Path(path).rglob("*.docx"):
        all_txt_files.append(file.parent / file.name)

    result = []
    for txt_file in all_txt_files:
        doc = docx2python(txt_file)
        for i in range(len(doc.body[0][0][0])):
            if(doc.body[0][0][0][i] != '' and doc.body[0][0][0][i] != ' ' and len(doc.body[0][0][0][i].split(" ")) > 4):
                result.append([doc.body[0][0][0][i], label])
    
    return result
```

```python
theme_all = read_docx_folder('Corpus/Theme/different', 2)
```

```python
theme_df = pd.DataFrame(theme_all, columns = ['text', 'label'])
theme_df
```
