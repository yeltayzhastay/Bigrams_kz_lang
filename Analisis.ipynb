```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
df = pd.read_csv('Corpus/big_corpus_cleaned.csv')
df
```

```python
cv = CountVectorizer()
X = cv.fit_transform(df.loc[df['label'] == 2].text)

tfidf = TfidfVectorizer()
Y = tfidf.fit_transform(df.loc[df['label'] == 2].text)

features = (cv.get_feature_names())
sums = X.sum(axis = 0)

sumy = Y.sum(axis = 0)
data1 = []
for col, term in enumerate(features):
    data1.append( (term, (sums[0,col] * sumy[0, col])))
ranking = pd.DataFrame(data1, columns = ['term','value'])
words = (ranking.sort_values('value', ascending = False))
```
